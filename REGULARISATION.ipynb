L1 Regularization (Lasso penalisation)
The L1 regularization adds a penalty equal to the sum of the absolute value of the coefficients. 

The L1 regularization will shrink some parameters to zero. Hence some variables will not play any role in the model, L1 regression can be seen as a way to select features in a model. Let’s see this with an example!

# where to use l1 or l2

On the one hand, L1 wants errors to be all or nothing, which leads to sparse weights since all the errors are concentrated in one place. As a simple example, let's say there are six weights, then L1 = |0|+|0|+|0|+|-3|+|0|+|1.8| = 4.8, or alternatively, L1 = |0.7|+|1.1|+|-0.6|+|0.9|+|0.7|+|-0.8| = 4.8. For L1 Norm, they both look about the same, and generally will lean towards the former.

#On the other hand, L2 wants all errors to be tiny and heavily penalizes anyone who doesn't obey. Looking back the at the our example, L2 = sqrt(0^2+0^2+0^2+3^2+0^2+1.8^2) = 3.5, or alternatively, L2 = sqrt(0.72+1.12+(-0.6)2+0.92+0.7+(-0.8)2) = 2. In this case, the L2 norm would clearly prefer the dispersed smaller values in the latter distribution.

In summary, L1 = sparse weights, L2 = small distributed weights.

```python
import numpy
numpy.sqrt(0**2+0**2+0**2+3**2+0**2+1.8**2)
```

In particular, consider the vectors a=(0.5,0.5) and b=(−1,0).  We can compute two possible norms:

∥a∥1=|0.5|+|0.5|=1 and ∥b∥1=|−1|+|0|=1


∥a∥2=0.52+0.52−−−−−−−−−√=1/2–√<1 and ∥b∥2=(−1)2+(0)2−−−−−−−−−−√=1

So, the two vectors are equivalent with respect to the L1 norm but different with respect to the L2 norm.  This is because squaring a number punishes large values more than it punishes small values.
