```python
# get K  using Decision trees
from sklearn import tree
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
def GetKFromDT( i,x ):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(x)
    cl =kmeans.labels_
    y=cl
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(x, y)
    YDT=clf.predict(x)
    accuracy =accuracy_score(y, YDT)
    return accuracy

bestAcc=0
bestK=0
for i in range(3,int(x.size/100)):
    acc=GetKFromDT(i,x)
    if bestAcc<acc:
        bestAcc=acc
        bestK=i
print (bestK)
print (acc)
kmeans = KMeans(n_clusters=bestK ,random_state=0).fit(x)
df['class']= kmeans.labels_
```

```python
import pandas as pd
import pandas
import scipy
import numpy
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
from sklearn.metrics import accuracy_score

df=pd.read_excel('C:\\Users\\Dell-pc\\Desktop\\Data science ML DL\\Practice\\Dataset\\data set important\\titanic.xls')
df.head()

for c in df.columns:
    df[c]=df[c].apply(str)
    le=preprocessing.LabelEncoder().fit(df[c])
    df[c] =le.transform(df[c])
    

y=df['survived']
x=df.loc[0:, df.columns != 'survived'] 
#Import Library 
from sklearn.ensemble import GradientBoostingClassifier 
#Assumed you have, X (predictor) and Y (target) for
#training data set and x_test(predictor) of test_dataset #Create Gradient Boosting Classifier object 
model= GradientBoostingClassifier(n_estimators=10,learning_rate=1, max_depth=1, random_state=0) 
#Train the model using the training sets and check score 
model.fit(x, y) 
#Predict Output 
predicted= model.predict(x)

accuracy =accuracy_score(y, predicted)
accuracy*100

# accuracy decrease when increase in learning rate for same max_depth and  number of estimators
#accuracy increase when decrease in learning rate for same max_depth and  number of estimators

#accuracy remain almost same when increase in n_estimators  for same max_depth and number of estimators
#accuracy decrease when decrease in n_estimators  for same max_depth and number of estimators

#accuracy increase when increase in max_depth for same n_estimators and  learning rate
#accuracy decrease when decrease in max_depth for same n_estimators and  learning rate




```
